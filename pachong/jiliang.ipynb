```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.stats import chisquare,chi2
```

```python
 data=pd.read_csv("C:\\Users\\hasee\\workspace\\workspace\\pachong\\houseprice4.csv")
```

```python
data
```

```python
X_train=data[['income','landprice','buildprice','rate','urbanlevel']]
```

```python
constant=pd.DataFrame(np.ones((14,)),columns={"constant"})
```

```python
constant
```

```python
X_train=pd.concat([X_train,constant],axis=1)
```

```python
X_train
```

```python
y_train=data['price']
```

```python
chisquare(X_train)
```

```python
chi2.fit(X_train)
```

```python
model_1=sm.OLS(y_train,X_train)
results=model_1.fit()
```

```python
print(results.summary())
```

```python
np.corrcoef(X_train.T)
```

```python
model_2=sm.WLS(y_train,X_train,weights=0.01)
results=model_2.fit()
```

```python
print(results.summary())
```
