# 1. Activition function

```python
import numpy as np

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedShuffleSplit
```

```python
def logit(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return np.maximum(0,x)

def derivative(f,x,epz=0.00000001):
    return (f(x+epz)-f(x-epz))/(2*epz)
```

```python
x=np.linspace(-5,5,200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(x,logit(x),'r-',label='logit')
plt.plot(x,relu(x),'b-',label='relu')
plt.plot(x,np.tanh(x),'g-',label='tanh')
plt.grid(True)
plt.legend()
plt.title("Activation function",fontsize=8)
#plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(x,derivative(logit,x),'r-',label='logit')
plt.plot(x,derivative(relu,x),'b-',label='relu')
plt.plot(x,derivative(np.tanh,x),'g-',label='tanh')
plt.grid(True)
plt.legend()
plt.title("derivative function",fontsize=8)
```

# 2. FNN for minst

** first we use tf.learn **

```python
import tensorflow as tf
from sklearn.datasets import fetch_mldata
mnist= fetch_mldata("MNIST original")
```

then import the minst dataset

```python
X,y=mnist['data'],mnist['target']
split=StratifiedShuffleSplit(n_splits=1,test_size=0.1,random_state=42)
for train_index,test_index in split.split(X,y):
    X_train=X[train_index]
    y_train=y[train_index].astype('int')
    X_test=X[test_index]
    y_test=y[test_index].astype('int')
    
```

```python
feature_columns=tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
```

```python
feature_columns
```

```python
dnn_clf=tf.contrib.learn.DNNClassifier(hidden_units=[300,100],n_classes=10,\
                                       feature_columns=feature_columns)
```

```python
dnn_clf.fit(X_train,y_train,batch_size=50,steps=40000)
```

```python
dnn_clf.evaluate(X_test,y_test)
```

```python
def plot_digit(X):
    some_digit_image=X.reshape(28,28)
    plt.imshow(some_digit_image,cmap = matplotlib.cm.binary)
    #plt.axis("off") #close the axis number
    plt.show()
    
some_digit=X_test[4]
plot_digit(some_digit)
some_digit=some_digit.reshape(len(some_digit),1)
```

then we try to use tf.plain to build a tensorflow

```python
n_inputs=28*28
n_hidden1=300
n_hidden2=100
n_outputs=10

X=tf.placeholder(tf.float32,shape=(None,n_inputs),name="X")
y=tf.placeholder(tf.int64,shape=(None),name="y")

def neuron_layers(X,n_neurons,name,activition=None):
    with tf.name_scope(name):
        n_inputs=int(X.get_shape()[1])
        stddev=2/np.sqrt(n_inputs)
        init=tf.truncated_normal((n_inputs,n_neurons),stddev=stddev)
        W=tf.Variable(init,name='weights')
        b=tf.Variable(tf.zeros([n_neurons]),name='bias')
        z=tf.matmul(X,W)+b
        if activition == 'relu':
            return tf.nn.relu(z)
        else:
            return z
```

```python
#also can use the fully_connected
with tf.name_scope("dnn"):
    hidden1=neuron_layers(X,n_hidden1,"hidden1",activition='relu')
    hidden2=neuron_layers(hidden1,n_hidden2,"hidden2",activition='relu')
    logit=neuron_layers(hidden2,n_outputs,"logit")
    
```

```python
with tf.name_scope("loss"):
    xentropy=tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y,logits=logit)
    #apply the softmax and calculate the cross the entropy
    loss=tf.reduce_mean(xentropy,name="loss")
```

```python
learning_rate=0.01
with tf.name_scope("train"):
    optimizer=tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
    training_op=optimizer.minimize(loss)
```

```python
with tf.name_scope("evaluate"):
    correct=tf.nn.in_top_k(logit,y,1)#in_top_k to assert your prediction and your label equal or not
    #return bool value
    accuracy=tf.reduce_mean(tf.cast(correct,tf.float32))#change the type of 'correct' into float32
```

```python
init=tf.global_variables_initializer()
saver=tf.train.Saver()
```

```python
n_epoch=400
batch_size=50
X,y=mnist['data'],mnist['target']
split=StratifiedShuffleSplit(n_splits=1,test_size=0.05,random_state=42)
for train_index,test_index in split.split(X,y):
    X_train=X[train_index]
    y_train=y[train_index].astype('int')
    X_test=X[test_index]
    y_test=y[test_index].astype('int')
    
y_train=y_train.reshape(len(y_train),1)
y_test=y_test.reshape(len(y_test),1)
```

```python

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epoch):
        for iteration in range(X_train.shape[0]//50):
            X_batch=X_train[iteration*batch_size:(iteration+1)*batch_size-1,:]
            y_batch=y_train[iteration*batch_size:(iteration+1)*batch_size-1,:]
            sess.run(training_op,feed_dict={X:X_batch,y:y_batch})
        acc_tr=accuracy.eval(feed_dict={X:X_train,y:y_train})
        acc_te=accuracy.eval(feed_dict={X:X_test,y:y_test})
        print(epoch,"Train accuracy:",acc_tr,"Test accuracy:",acc_te)
```
