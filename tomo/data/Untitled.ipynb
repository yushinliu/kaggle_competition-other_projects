```python
def Selectfeatures(input_set,name_set,lng_set,lat_set):
    for index in range(len(input_set)):
        
        if (input_set["type"][index].find("地名地址信息")==-1) and (input_set["type"][index].find(";住宅区;")==-1) and (input_set["type"][index].find(";临街院门;")==-1):
            continue                   
        else:
            name_set.append(input_set["name"][index])
            lng_set.append(input_set["lng"][index])
            lat_set.append(input_set["lat"][index])

           
```

```python
name_set=[]
lng_set=[]
lat_set=[]
Selectfeatures(sh_poi,name_set,lng_set,lat_set)
```

```python
new_poi=pd.DataFrame(name_set,columns=['name'])
new_poi.insert(1,'lng',lng_set)
new_poi.insert(2,'lat',lat_set)
```

```python
def Comparefeatures(household, new_poi):
    for index_1 in range(len(household)):
        if household["居住的小区"][index_1].find('-')==-1:
            for index_2 in range(len(new_poi)): 
                if household["居住的小区"][index_1] in new_poi["name"][index_2]:
                    household["address"][index_1]=new_poi["name"][index_2]
                    household["lat"][index_1]=new_poi["lat"][index_2]
                    household["lng"][index_1]=new_poi["lng"][index_2]
                    break
                else:
                    continue                                
        else: 
            temp=household["居住的小区"][index_1].split('-')
            for index_3 in range(2):
                if "路" in temp[index_3]:
                    temp[index_3]=temp[index_3].split("路")[0]
                else:
                        continue
            for index_2 in range(len(new_poi)): 
                if (temp[0] in new_poi["name"][index_2]) and (temp[1] in new_poi["name"][index_2]):
                    household["address"][index_1]=new_poi["name"][index_2]
                    household["lat"][index_1]=new_poi["lat"][index_2]
                    household["lng"][index_1]=new_poi["lng"][index_2]
                    break
                else:
                    continue                                
                    
            
```
