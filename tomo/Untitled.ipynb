```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
math.pi
```

```python
def gaussian(x,sigma,mu):
    value=(1/sigma*np.sqrt(2*math.pi))*np.exp(-(x-mu)**2/(2*sigma**2))
    return value
```

```python
x=np.linspace(-100,100,10000)
plt.figure(figsize=(14,8))
plt.plot(x,gaussian(x,30,0),color='black')
plt.axvline(0,color='black',label='$\hat{B}$',linewidth=1)
plt.axvline(-50,color='blue',label='$\dot{B}$',linewidth=1)
plt.axvline(-30,color='blue',label='$\dot{B} \pm \hat{B}*\sigma$',linestyle='--',linewidth=1)
plt.axvline(-70,color='blue',linestyle='--',linewidth=1)
plt.axis('off')
plt.legend()
plt.savefig('tomo_fig.jpg')
plt.show()
```
